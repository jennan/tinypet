```python
import numpy as np
import xarray as xr

from tinypet.data import XarraySource, Seq
import tinypet.operations as ops
import tinypet.sampler as sp
```

```python
era5 = xr.open_zarr('gs://weatherbench2/datasets/era5/1959-2023_01_10-6h-64x32_equiangular_conservative.zarr')
```

```python
era5
```

```python
source = XarraySource(era5, dim="time")
```

```python
source["2021-01-01T06"]
```

```python
select = ops.Select(["2m_temperature", "geopotential"])
level = ops.Sel(level=[50, 100, 150, 200])
compute = ops.Compute(scheduler='single-threaded')
mean = ops.Mean(dim="level")
rename = ops.Rename({"latitude": "lat", "longitude": "lon"})
stack = ops.ToStackedArray("feats", sample_dims=["lat", "lon"])
```

```python
pipeline = stack(rename(mean(compute(level(select(source))))))
pipeline["2021-01-01T06"]
```

```python
pipeline = source | select | level | compute | mean | rename | stack
pipeline["2021-01-01T06"]
```

```python
train_dates = sp.DateRange(end="20191231")
shuffle = sp.Shuffle(seed=1234)

train_pipeline = pipeline | train_dates | shuffle
train_pipeline.index
```

```python
max(train_pipeline.index)
```

```python
train_pipeline.index
```

```python
virtual_pipeline = select | level | compute | mean | rename | stack
virtual_pipeline
```

```python
new_pipeline = source | virtual_pipeline
new_pipeline["2021-01-01T06"]
```

```python
undo_stack = pipeline.undo()
(pipeline | undo_stack)["2021-01-01T06"]
```

```python
undo_stack_and_rename = pipeline.undo(2)
(pipeline | undo_stack_and_rename)["2021-01-01T06"]
```
