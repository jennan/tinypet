```python
import xarray as xr

from tinypet.data import XarraySource
import tinypet.operations as ops
```

```python
era5 = xr.open_zarr('gs://weatherbench2/datasets/era5/1959-2023_01_10-6h-64x32_equiangular_conservative.zarr')
```

```python
era5
```

```python
source = XarraySource(era5, dim="time")
```

```python
len(source)
```

```python
source.index
```

```python
next(iter(source))
```

```python
source["2021-01-01T06"]
```

```python
rename = ops.Rename({"latitude": "lat", "longitude": "lon"})
mean = ops.Mean(dim="level")
```

```python
pipeline = mean(rename(source))
pipeline["2021-01-01T06"]
```

```python
pipeline = source | rename | mean
pipeline["2021-01-01T06"]
```

```python
import tinypet.sampler as sp
```

```python
train_dates = sp.DateRange(end="20191231")
shuffle = sp.Shuffle(seed=1234)

train_pipeline = pipeline | train_dates | shuffle
train_pipeline.index
```
